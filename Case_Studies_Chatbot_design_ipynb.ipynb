```python
import re
import random
import sympy
# ============================================
# Intent & Response Setup
# ============================================

INTENTS = {
    "greeting": {
        "patterns": [r"hi", r"hello", r"hey", r"good morning"],
        "responses": ["Hello! How can I help you today?", "Hi there! Ready to learn?"]
    },
    "math_practice": {
        "patterns": [r"math", r"practice", r"arithmetic"],
        "responses": ["Let's practice math! Solve this: {problem}", None]  # None triggers problem generation
    },
    "python_help": {
        "patterns": [r"python", r"programming", r"syntax"],
        "responses": [
            "In Python, lists are mutable, tuples are immutable.",
            "Use 'def' to define a function: 'def my_func():'."
        ]
    },
    "exit": {
        "patterns": [r"bye", r"quit", r"exit"],
        "responses": ["Goodbye! Come back anytime.", "See you later!"]
    }
}

# ============================================
# Core Functions
# ============================================

# Removes special characters and lowercases text for consistent matching.
def clean_input(text):
    """Preprocess user input by removing case and special characters."""
    return re.sub(r"[^a-zA-Z0-9 ]", "", text.lower())

# Matches cleaned input to intents using regex.
def get_intent(user_input):
    """Match user input to predefined intents using keywords."""

    cleaned_input = clean_input(user_input)
    for intent, data in INTENTS.items():
        for pattern in data["patterns"]:
            if re.search(pattern, cleaned_input):
                return intent
    return None

# Function to generate a simple math problem and answer using sympy
def generate_math_problem():
    """Generates a simple addition problem."""

    x = sympy.randprime(1, 10)  # Generate a random prime number between 1 and 10
    y = sympy.randprime(1, 10)  # Generate another random prime number between 1 and 10
    problem = f"{x} + {y}"
    answer = x + y
    return problem, answer

def handle_math_answer(correct_answer):
    """Check if the user's answer matches the expected value."""
    try:
        user_answer = float(input("Your answer: "))
        return "Correct!" if user_answer == correct_answer else f"Oops! The answer was {correct_answer}."
    except ValueError:
        return "Please enter a number."
def main():
    current_state = None  # Tracks if the user is in a practice session
    print("TutorBot: Hello! Ask about Python or type 'practice math'.")

    while True:
        user_input = input("You: ").strip()  # Capture all input here

        # Check if we're waiting for a math answer
        if current_state and current_state.get("state") == "awaiting_math_answer":
            try:
                user_answer = float(user_input)
                correct_answer = current_state["answer"]
                if user_answer == correct_answer:
                    print("TutorBot: Correct!")
                else:
                    print(f"TutorBot: Oops! The answer was {correct_answer}.")
            except ValueError:
                print("TutorBot: Please enter a number.")
            current_state = None
            continue  # Skip intent processing after handling the answer

        # Now back to normal intent handling
        intent = get_intent(user_input)

        if not intent:
            print("TutorBot: I'm not sure. Try 'practice math' or 'Python help'.")
            continue

        responses = INTENTS[intent]["responses"]
        response = random.choice(responses)

        if intent == "math_practice":
            problem, answer = generate_math_problem()
            print(f"TutorBot: {problem}")
            current_state = {"state": "awaiting_math_answer", "answer": answer}

        elif intent == "exit":
            print(f"TutorBot: {response}")
            break

        else:
            print(f"TutorBot: {response}")


if __name__ == "__main__":
    main()
```
